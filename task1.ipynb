```python
import matplotlib.pyplot as plt
import numpy as np
ages = np.random.normal(30, 10, 1000) 
genders = ['Male', 'Female', 'Others']*333 + ['Male'] 
plt.figure(figsize=(10, 5))
unique, counts = np.unique(genders, return_counts=True)
plt.bar(unique, counts, color='yellow', edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
